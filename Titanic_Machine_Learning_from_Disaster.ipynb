***Missing Values in Data-sets***

```python
print('Missing values in training set: ')
print(train.isnull().sum())
print('\nMissing values in test set: ')
print(test.isnull().sum())
```

***Duplicates in Data-sets***

```python
if train.duplicated().sum() == 0:
  print('Duplicates in training set : 0')
else:
  print('Duplicates in training set :' + str(train.duplicated().sum()))

if test.duplicated().sum() == 0:
  print('Duplicates in test set : 0')
else:
  print('Duplicates in test set :' + str(test.duplicated().sum()))
```
